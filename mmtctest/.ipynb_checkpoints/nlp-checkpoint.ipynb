```python
#pre process connecting trains
lev1 = []
for i in range(0,tr):
    if train.iloc[i,12] == s.upper():
        if train.iloc[i,7] in tsource and train.iloc[i,7] != s.upper():
            c = list(str(train.iloc[i,5]))
            if date in c:
                dist = train.iloc[i,8]
                if dist < 50:
                    time = 1
                else:
                    time = int(dist/50)
                print("time = ", time)
                hr,_,_ = list(map(int,train.iloc[i,14].split(':')))
                rtime = time + hr
                if rtime > 23:
                    label = int(rtime/24) * 1000 
                    print("label = ", label)
                else:
                    label = rtime
                    print("label = ", label)
                d2 = train.iloc[i,7]
                tnum = train.iloc[i,15]
                l = (d2,label,tnum,i)
                lev1.append(l)
    
```
